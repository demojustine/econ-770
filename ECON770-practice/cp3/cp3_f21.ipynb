In this exercise, you will create three time series plots of the number of banks, number of offices, and number of offices per bank between 1934 and 2017. 

### Part (a):

1. Open your notebook from coding practice \#2. Run your code for exercise 2. You should have a DataFrame named `banking_data`.

```python
#Exercise 1
#part (a)
import pandas as pd

banking_data = pd.read_csv('banking_data.csv')
banking_data.head(2)
```

### Part (b):

1. Create a figure with three subplots (all in a single column). Set the figure size to (15, 15).
2. In the first subplot, create a line plot of the number of banks by year. Make the line dotted and black. Label the y-axis 'Number of banks'. Set the title of this subplot as 'Number of U.S. commercial banks, 1934-2017'.
3. In the second subplot, create a line plot of the number of offices by year. Make the line dashed and red. Label  the y-axis 'Number of offices'. Set the title of this subplot as 'Number of offices of U.S. commercial banks, 1934-2017'.
4. In the third subplot, create a line plot of the average number of offices per bank. Make the line blue with a dash-dot pattern. Label the  y-axis 'Number of offices per bank'. Set the title of this subplot as 'Average number of offices per bank, 1934-2017'.

```python
#part (b)
#1
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3,1, figsize=(15,15))
# the number of banks by year
ax[0].plot(banking_data['year'], banking_data['banks'], linestyle = 'dotted', linewidth = 2.0,  color = 'black')
ax[0].set_title('Number of U.S. commercial banks, 1934-2017', fontsize = 20)
ax[0].set_ylabel('Number of banks', fontsize = 16)

# the number of offices by year
ax[1].plot(banking_data['year'], banking_data['offices'], linestyle = 'dashed', linewidth = 2.0,  color = 'red')
ax[1].set_title('Number of offices of U.S. commercial banks, 1934-2017', fontsize = 20)
ax[1].set_ylabel('Number of offices', fontsize = 16)

#  the average number of offices per bank
ax[2].plot(banking_data['year'], banking_data['offices_per_bank'], linestyle = 'dashdot', linewidth = 2.0,  color = 'blue')
ax[2].set_title('Average number of offices per bank, 1934-2017', fontsize = 20)
ax[2].set_ylabel('Number of offices per bank', fontsize =16)

plt.show()
```

```python
#part(a)
#1
import pandas_datareader.data as web   
import datetime as dt 

start = dt.datetime(1950,1,1)
gdp = web.DataReader('GDPC1', 'fred' , start)

gdp.head(2)
```
